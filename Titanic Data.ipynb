```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename='train.csv'
titanic_df = pd.read_csv(filename)
titanic_df
```

```python
# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Check for missing values
print(titanic_df.isnull().sum())
```

```python
# Impute missing values (for simplicity, we'll fill missing values with median/mean)
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
titanic_df['Fare'].fillna(titanic_df['Fare'].mean(), inplace=True)
titanic_df['Cabin'].fillna('Unknown', inplace=True)

# Display the updated DataFrame
print(titanic_df)
```

```python
filename='train.csv'
titanic_df = pd.read_csv(filename)
titanic_df
```

```python
# Drop irrelevant columns
titanic_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)

# Display the updated DataFrame
print(titanic_df)
```

```python
# Convert categorical variables to numerical
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked', 'Pclass'], drop_first=True)
print(titanic_df)
```

```python
# Descriptive statistics
print(titanic_df.describe())
```

```python
# Visualize the survival count
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.show()
```

```python
# Explore survival rate based on Pclass
sns.barplot(x='Pclass_2', y='Survived', data=titanic_df)
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
# Explore survival rate based on Sex
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Explore survival rate based on Age
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True)
plt.title('Survival by Age')
plt.show()
```

```python
# Explore correlations between numerical variables
numeric_columns = titanic_df.select_dtypes(include=[np.number]).columns
correlation_matrix = titanic_df[numeric_columns].corr()

# Plot the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Handling Outliers in 'Fare'
sns.boxplot(x='Fare', data=titanic_df)
plt.title('Boxplot of Fare')
plt.show()
```

```python
# Apply log transformation to 'Fare' to mitigate the impact of outliers
titanic_df['Fare'] = np.log1p(titanic_df['Fare'])
print(titanic_df)
```

```python
# Feature Engineering - Family Size
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
print(titanic_df)
```

```python
# Age Distribution
sns.histplot(x='Age', data=titanic_df, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
# Fare Distribution after log transformation
sns.histplot(x='Fare', data=titanic_df, kde=True)
plt.title('Distribution of Fare (after log transformation)')
plt.show()
```

```python
# Survival by Age Group
bins = [0, 12, 18, 30, 50, 100]
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=bins, labels=labels)
sns.barplot(x='AgeGroup', y='Survived', data=titanic_df, order=labels)
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
# Drop or encode 'Cabin' column
titanic_df.drop(['Cabin'], axis=1, inplace=True)

# Explore correlations between numerical variables
numeric_columns = titanic_df.select_dtypes(include=[np.number]).columns
correlation_matrix = titanic_df[numeric_columns].corr()

# Plot the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
